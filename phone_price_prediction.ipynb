# Görselleştirmeler

```python
plt.figure(figsize=(21, 9))
corr_matrix = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, cmap='Greens', annot=True, mask=mask)
plt.title('Korelasyon Haritası')
```

```python
plt.figure(figsize=(21,9))
plt.xlabel("Telefon Markaları", labelpad=30, fontsize=16)
plt.title("Telefon Marka Adedi", pad=30, fontsize=18)
plt.ylabel("Adet", labelpad=30, fontsize=16)
sns.countplot(x = df.brand, palette="pastel", edgecolor="black")
```

```python
plt.figure(figsize=(21,9))
plt.xlabel("Adet", labelpad=30, fontsize=16)
plt.ylabel("İşletim Sistemi", labelpad=30, fontsize=16)
sns.countplot(y = df.os, palette="Set1", edgecolor="#100")
```

```python
df.hist(figsize=(20, 15), bins=30, edgecolor='black')
plt.tight_layout()
```

```python
category_counts = df['battery_type'].value_counts()
explode = (0.2, 0)
plt.figure(figsize=(21, 9))
plt.pie(category_counts, explode=explode,shadow=True, labels=category_counts.index, autopct='%1.1f%%', startangle=180, colors=["#1778c2","#8c2108"], labeldistance=1.1)
plt.title('Batarya Tip Dağılımı')
```

```python
categorical = df.select_dtypes(include = ['bool'])

plt.figure(figsize=(15, 20))
for i, col in enumerate(categorical.columns):
    plt.subplot(5, 2, i + 1)
    sns.countplot(data=categorical, x=col, palette='Set2')
    plt.title(f'{col}')
    plt.xticks(rotation=45)

plt.tight_layout()
```

# Aykırı Verileri Temizleme

```python
Q1 = df['battery'].quantile(0.25)
Q3 = df['battery'].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[~((df['battery'] < (Q1 - 1.5 * IQR)) | (df['battery'] > (Q3 + 1.5 * IQR)))]
```

```python
Q1 = df['weight(g)'].quantile(0.25)
Q3 = df['weight(g)'].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[~((df['weight(g)'] < (Q1 - 1.5 * IQR)) | (df['weight(g)'] > (Q3 + 1.5 * IQR)))]
```

```python
Q1 = df['price(USD)'].quantile(0.25)
Q3 = df['price(USD)'].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[~((df['price(USD)'] < (Q1 - 1.5 * IQR)) | (df['price(USD)'] > (Q3 + 1.5 * IQR)))]
```
